```python
def savedmasked(tests,masks,folder):
    for i in range(len(tests)):
        nametest = tests[i].split('/')[-1]
        namemask = masks[i].split('/')[-1]
        #print(nametest,namemask)
        if(nametest == namemask):
            print(nametest, namemask)
            src = cv.imread(masks[i])
            graysrc = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
            ret, binary = cv.threshold(graysrc , 0, 255, cv.THRESH_BINARY | cv.THRESH_TRIANGLE)
            normalized = (binary - binary.min()) / (binary.max() - binary.min())
            imgsm = cv.imread(tests[i])
            graysm = cv.cvtColor(imgsm, cv.COLOR_BGR2GRAY)
            smmasked = normalized*graysm
            plt.imshow(smmasked)
            plt.axis('off')
            plt.savefig(f'results/{folder}/{nametest}',dpi=169.8,bbox_inches='tight',pad_inches=0)
            #plt.savefig('prueba1.png',dpi=169.8,bbox_inches='tight',pad_inches=0)
            #plt.show()
            plt.close()
        #break
```

```python
def savedmasked_2(tests,masks,folder):
    for i in range(len(tests)):
        nametest = tests[i].split('/')[-1]
        namemask = masks[i].split('/')[-1]
        #print(nametest,namemask)
        if(nametest == namemask):
            print(nametest, namemask)
            src = cv.imread(masks[i])
            graysrc = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
            ret, binary = cv.threshold(graysrc , 0, 255, cv.THRESH_BINARY | cv.THRESH_TRIANGLE)
            normalized = (binary - binary.min()) / (binary.max() - binary.min())
            
            imagen = Image.open(tests[i])
            # Convierte la imagen a escala de grises
            imagen_gris = imagen.convert('L')

            # Invierte la escala de grises
            imagen_invertida = ImageOps.invert(imagen_gris)
            #print(type(normalized))
            #print(type(imagen_invertida))
            smmasked = normalized*imagen_invertida
            
            smmasked = Image.fromarray(smmasked.astype('uint8'))
            # Guarda la imagen invertida
            smmasked.save(f'results/detect/{folder}/{nametest}')
        #break
```
